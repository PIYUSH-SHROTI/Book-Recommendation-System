```python
z=q[q].index
```

```python
fr=r[r['Book-Title'].isin(z)]
```

```python
fr
```

```python
pt=fr.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pt
```

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
ss= cosine_similarity(pt)
```

```python
ss.shape
```

```python
def recommend(name):
    ind = np.where(pt.index==name)[0][0]
    item = sorted(list(enumerate(ss[ind])),key=lambda x:x[1],reverse=True)[1:6]
    
    for i in item:
        print(pt.index[i[0]])
```

```python
recommend('Year of Wonders')
```
